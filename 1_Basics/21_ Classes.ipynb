# Classes

```python
class LukesList:
    def __init__(self):
        """Initialize an empty list."""
        self._items = []

    def add(self, item):
        """Add an item to the end of the list."""
        self._items.append(item)

    def __getitem__(self, index):
        """Retrieve an item by index."""
        return self._items[index]

    def __setitem__(self, index, value):
        """Set an item at a specific index."""
        self._items[index] = value

    def __repr__(self):
        """Return a string representation of the list."""
        return str(self._items)

    def __len__(self):
        """Return the length of the list."""
        return len(self._items)
```

```python
my_list = LukesList()
```

```python
# We usually use camel-case for class names
# init must start with self, but it doesn't have to be called self (pay attention whe assigning values to the attributes)
class BaseSalary:
  def __init__(self, base_salary, bonus_rate=.1, symbol='$'):
    self.base_salary = base_salary
    self.bonus_rate = bonus_rate
    self.symbol = symbol
    self.total_salary = base_salary * (1 + bonus_rate)
    self.bonus = self.total_salary - base_salary

  def __repr__(self):
    return f'{self.symbol}{self.base_salary:,.0f}'

  def show_salary(self):
    return f'{self.symbol}{self.total_salary:,.0f}'

  def show_bonus(self):
    return f'{self.symbol}{self.bonus:,.0f}'
```

```python
salary = BaseSalary(100000)
```

```python
salary.show_salary()
```

```python
salary.show_bonus()
```
