## User-Defined Functions

```python
base_salary = 100000
bonus_rate = 0.1

total_salary = base_salary * (1 + bonus_rate)
total_salary
```

```python
base_salary = 100000
bonus_rate = 0.1

def calculate_salary(base_salary, bonus_rate):
  return base_salary * (1 + bonus_rate)

total_salary = calculate_salary(base_salary, bonus_rate)

total_salary
```

We can add optional arguments by having a `=` sign after the argument with a default value:

```python
def calculate_salary(base_salary, bonus_rate=0.1):
  return base_salary * (1 + bonus_rate)
```

```python
total_salary = calculate_salary(base_salary, 0.5)

total_salary
```

```python
total_salary = calculate_salary(base_salary)

total_salary
```
