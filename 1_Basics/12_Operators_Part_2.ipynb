## Logical Operators

| Operator | Name     | Example |
|----------|----------|---------|
| and      | Logical AND | x and y |
| or       | Logical OR  | x or y  |
| not      | Logical NOT | not x   |

```python
job_work_from_home = True
job_health_insurance = False
# job_no_degree_mentioned = True

print(job_work_from_home and job_health_insurance)
print(job_work_from_home or job_health_insurance)
```

## Identity Operators

| Operator | Name               | Example   |
|----------|--------------------|-----------|
| is       | Identity           | x is y    |
| is not   | Not identity       | x is not y|

Verifies if two variables are using the same memory location for the value they are holding.

```python
salary_kelly = 100_000
salary_luke = 100_000

print(salary_kelly == salary_luke)
print(salary_kelly is salary_luke)
```

```python
core_skills = ['Python', 'R', 'SQL']
lukes_skills = core_skills
```

```python
lukes_skills is core_skills
```
