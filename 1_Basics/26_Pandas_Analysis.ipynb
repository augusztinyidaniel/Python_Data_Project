```python
from datasets import load_dataset
import pandas as pd

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
```

```python
df.describe()
```

```python
df.info()
```

There are a lot of other methods we can run to understand the key statistics of different columns in our dataframe:
- `df.count()` - count of values
- `df.sum()` - sum of values
- `df.cumsum()` - cummulative sum of values
- `df.min()` / `df.max()` - indexes of min/max values
- `df.mean()` - mean of values
- `df.median()` - median of values
- `df.mode()` - mode of values
- `df.column_name.value_counts()` - unique counts

```python
df.count()
```

```python
df.median()
```
