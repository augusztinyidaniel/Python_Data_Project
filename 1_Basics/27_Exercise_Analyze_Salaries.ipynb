```python
from datasets import load_dataset
import pandas as pd

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
df['job_posted_date'] = pd.to_datetime(df.job_posted_date)
```

```python
df['job_country'].value_counts().head(20)
```

```python
df['job_country'].isin(['Hungary']).any()
```

```python
hun_jobs = df[df['job_country'] == 'Hungary']
```

```python
hun_jobs
```

```python
hun_jobs = hun_jobs[hun_jobs['salary_year_avg'].notna()]
```

```python
hun_jobs.groupby('job_title_short')['salary_year_avg'].count()
```

```python
hun_jobs.groupby('job_title_short')['salary_year_avg'].min()
```

```python
hun_jobs.groupby('job_title_short')[['salary_year_avg']].min()
```

```python
hun_jobs.groupby('job_title_short')['salary_year_avg'].agg(['median', 'min', 'max', 'count']).sort_values(by='median', ascending=False)
```
