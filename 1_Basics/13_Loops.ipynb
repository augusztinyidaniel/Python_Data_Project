# Loops

- `for` loops
- `while` loops

## `for` loops

```python
numbers = [0, 1, 2, 3, 4, 5]

for number in numbers:
    print(number)
```

```python
characters = 'PYTHON'

for character in characters:
  print(character)
```

```python
years_experience = {
    'Luke':  3,
    'Kelly': 6,
    'Ken':   4,
    'Alex':  3
}
```

Let's hire somebody with more than 5 years of experience!

```python
for person in years_experience:
  print(person)
```

```python
for value in years_experience.values():
  print(value)
```

```python
for key, value in years_experience.items():
  print(key, value)
```

```python
for key, value in years_experience.items():
  if value > 5:
    print(key, 'has', value, 'years of experience')
```

```python
for _ in range(100):
  print(_)
```

## `while` loops

```python
count = 1
while count < 5:
  print(count)
  count += 1
```

We only want to find the first occurence of the condition being met. (More than 5 years)

```python
years_experience = {
    'Luke':  3,
    'Kelly': 6,
    'Ken':   4,
    'Alex':  3,
    'Mary':  7
}
```

```python
years = 0
index = 0

years_list = list(years_experience.items())
while years < 5:
  key, years = years_list[index]
  if years >= 5:
    print(key, 'has', years, 'years of experience')
  index += 1
```
