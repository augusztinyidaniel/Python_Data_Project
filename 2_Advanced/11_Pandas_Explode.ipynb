```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt 

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
import pandas as pd

data = {
    'job_title_short': ['Data Analyst', 'Data Scientist', 'Data Engineer'],
    'job_skills': [['excel', 'sql', 'python'], ['python', 'r'], ['aws', 'python', 'airflow']]
}

df_skill_list = pd.DataFrame(data)

df_skill_list
```

If we want to count how many times for example excel appears, we would have to iterate through these lists. But, if we use `explode()`, every item in the list gets its own row.

```python
df_skill_list.explode('job_skills')
```

```python
df_skill_list.explode('job_skills').value_counts('job_skills')
```

```python
df_skill_list.explode('job_skills').value_counts('job_skills').plot(kind='bar')
```

```python
df_exploded = df.explode('job_skills')
```

```python
df_exploded['job_skills'].value_counts().head(10).plot(kind='bar')
```

```python
df_exploded.groupby(['job_title_short', 'job_skills']).size()
```

```python
skills_count = df_exploded.groupby(['job_title_short', 'job_skills']).size()

df_skills_count = skills_count.reset_index(name='skill_count')
df_skills_count
```

```python
df_skills_count = df_skills_count.sort_values(by='skill_count', ascending=False)

df_skills_count
```

```python
job_title = 'Data Analyst'
top_skills = 10

df_skills_final = df_skills_count[df_skills_count['job_title_short'] == job_title].head(top_skills)

df_skills_final.plot(kind='barh', x='job_skills', y='skill_count')
plt.gca().invert_yaxis()
plt.title(f'Top {top_skills} Skills for {job_title}')
plt.xlabel('Job Posting Count')
plt.ylabel('')
plt.legend().set_visible(False)
plt.show()
```
