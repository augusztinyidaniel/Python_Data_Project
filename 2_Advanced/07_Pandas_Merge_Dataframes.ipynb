```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt 

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

Creating the dataframes that are used in the demonstration in the course material:

```python
# Define the data
data = {
    'job_id': [1, 2, 3, 4, 5],
    'job_title': ['Data Scientist', 'Software Engineer', 'Product Manager', 'Marketing Director', 'HR Manager'],
    'company_name': ['Google', 'Microsoft', 'Apple', 'Nike', 'Starbucks'],
    'job_location': ['Mountain View, CA', 'Redmond, WA', 'Cupertino, CA', 'Beaverton, OR', 'Seattle, WA']
}

# Create the DataFrame
df_jobs = pd.DataFrame(data)

# Display the DataFrame
df_jobs
```

```python
# Define the data
data = {
    'company_name': ['Google', 'Microsoft', 'Apple', 'Nike', 'Starbucks'],
    'industry': ['Technology', 'Technology', 'Technology', 'Apparel', 'Food & Beverage'],
    'company_size': ['100,000+', '100,000+', '100,000+', '75,000+', '346,000+']
}

# Create the DataFrame
df_companies = pd.DataFrame(data)

# Display the DataFrame
df_companies
```

Now we can perform the merges demonstrated in the course material.

```python
df_jobs.merge(df_companies, on='company_name')
```

Let's move back to our original DataFrame and use the pivot table we built in the previous part but make it US so it fits the other data better.

```python
# Filtering for Hungary
df_US = df[df['job_country'] == 'United States'].copy()

# Getting a column with the month names
df_US['job_posted_month'] = df_US['job_posted_date'].dt.strftime('%B')

# Creating the pivot table
df_US_pivot = df_US.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')

# Resetting index
df_US_pivot.reset_index(inplace=True)

# Getting the numerical value of every month in a column
df_US_pivot['month_no'] = pd.to_datetime(df_US_pivot['job_posted_month'], format='%B').dt.month

# Sorting chronologically
df_US_pivot.sort_values('month_no', inplace=True)

# Making the now sorted months the index
df_US_pivot.set_index('job_posted_month', inplace=True)

# Dropping the no longer needed month numbers
df_US_pivot.drop(columns='month_no', inplace=True)

# Display
df_US_pivot
```

```python
# Importing the fictitious software developer data
df_US_software_pivot = pd.read_csv("https://lukeb.co/software_csv", index_col='job_posted_month')
df_US_software_pivot
```

Now let's get all our data in one DataFrame using the `merge()` method.

```python
df_US_merged = df_US_pivot.merge(df_US_software_pivot, on='job_posted_month')
df_US_merged
```

It is possible that we want to merge or join our data differently. For that we can use the `how` argument in the `merge()` method.

More about this on the [Cheat Sheet](https://pandas.pydata.org/Pandas_Cheat_Sheet.pdf), Combine Data Sets part.

We want to now plot the top 5 jobs based on count of jobs in the year and see how they trend over time.

```python
# Making a list of the top 5 jobs
top_5 = df_US_merged.sum().sort_values(ascending=False).head(5).index.to_list()

# If we see such a long series of methods and attributes, it is a good idea to break it up into lines using parentheses
top_5 = (
    df_US_merged
    .sum()
    .sort_values(ascending=False)
    .head(5)
    .index
    .to_list()
)

top_5
```

```python
df_US_merged[top_5].plot(kind='line')
```

```python
df_US_merged[top_5].plot(kind='line')
plt.title('Monthly Job Postings for Top Tech Jobs in the US')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.ylim(0, 20000)
plt.legend()
plt.show()
```
