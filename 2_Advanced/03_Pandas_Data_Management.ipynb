```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt 

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

This is good, however it is not exactly representistic of the dataset itself. It isn't a true sample. For that, we can use the `sample` method.

```python
# This way we get 10 random rows every time.
df.sample(10)
```

```python
# This way it is always the same 10 rows (42 is not important, could be any other number for another 10 rows)
df.sample(10, random_state=42)
```

# `copy()`

```python
dataset = load_dataset('lukebarousse/data_jobs')
df_original = dataset['train'].to_pandas()
```

```python
df_altered = df_original
```

```python
df_altered.loc[:5, 'salary_year_avg']
```

```python
median_salary = df_altered['salary_year_avg'].median()

median_salary
```

```python
df_altered['salary_year_avg'] = df_altered['salary_year_avg'].fillna(median_salary)

df_altered.loc[:5, 'salary_year_avg']
```

```python
df_original.loc[:5, 'salary_year_avg']
```

```python
print(id(df_original))
print(id(df_altered))
```

These variables, however their names are different, referencing the exact same dataframe.

```python
dataset = load_dataset('lukebarousse/data_jobs')
df_original = dataset['train'].to_pandas()
```

```python
df_altered = df_original.copy()
print(id(df_original))
print(id(df_altered))
```

```python
median_salary = df_altered['salary_year_avg'].median()

df_altered['salary_year_avg'] = df_altered['salary_year_avg'].fillna(median_salary)

df_altered.loc[:5, 'salary_year_avg']
```

```python
df_original.loc[:5, 'salary_year_avg']
```

As we can see, now the original dataset is untouched.
