# Exercise - Analyzing Data Science Job Demand

```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt 

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

Let's filter the orginal dataframe for the country of our interest, for example Hungary.

```python
df_HU = df[df['job_country'] == 'Hungary'].copy()

df_HU
```

Now we extract the month values but not in number form but rather the names of the months.

More on fotmat codes: [Documentation](https://docs.python.org/3/library/datetime.html#format-codes)

```python
df_HU['job_posted_month'] = df_HU['job_posted_date'].dt.strftime('%B')

df_HU
```

Now we need to pivot our data in order to get it into the format necessary to plot it. Specifically, we want the job month (along with the index for a specific reason) and then along the columns we are going to have the different job titles. We are aggregating the count of these different jobs over time.

```python
df_HU.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
```

As we can see, we now have the data almost in the correct format, but the order of the months is all wrong. We need that in chronological order.

```python
df_HU_pivot = df_HU.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')

# Resetting index (before this the month names are the index)
df_HU_pivot.reset_index(inplace=True)

# Creating a column with dates in it that contains the numerical value of each month
df_HU_pivot['month_no'] = pd.to_datetime(df_HU_pivot['job_posted_month'], format='%B')

df_HU_pivot['month_no']
```

```python
# Modifying our previous code to get just the number of the month
df_HU_pivot['month_no'] = pd.to_datetime(df_HU_pivot['job_posted_month'], format='%B').dt.month

df_HU_pivot['month_no']
```

Let's check on our dataframe.

```python
df_HU_pivot
```

Now we need to sort our values by the month number. 

```python
df_HU_pivot.sort_values('month_no', inplace=True)
```

We can now set the index.

```python
df_HU_pivot.set_index('job_posted_month', inplace=True)
```

```python
df_HU_pivot
```

It is now indexed by the month, in chronological order, now we can get rid of the `month_no` column.

```python
df_HU_pivot.drop(columns='month_no', inplace=True)
df_HU_pivot
```

We are ready to plot!

```python
df_HU_pivot.plot(kind='line')
```

It's a bit cluttered. Let's clean this up to only show the top 3 jobs.

```python
top_3 = df_HU['job_title_short'].value_counts().head(3)
top_3 = top_3.index.to_list()
top_3
```

```python
df_HU_pivot[top_3].plot(kind='line')
```

Let's improve our plot even further, with titles and labels.

```python
df_HU_pivot[top_3].plot(kind='line')
plt.title('Monthly Job Postings for Top 3 Data Jobs in Hungary')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.legend()
plt.show()
```
